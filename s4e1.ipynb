```python
df = pd.read_csv("/kaggle/input/playground-series-s4e1/train.csv")  # (165034, 14)
data_test = pd.read_csv("/kaggle/input/playground-series-s4e1/test.csv")
df.head()
```

```python
train_target = df['Exited']

df = df.drop(columns=['Exited'])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

def feature_processing(df):
    df = pd.get_dummies(df, columns=['Geography'], prefix = ['Geo'])
    df = df.drop(['id', 'CustomerId', 'Surname'], axis = 1)
    df['Gender'].replace({'Male': 0, 'Female': 1}, inplace=True)
    
    scaler = StandardScaler()
    df_scaled = scaler.fit_transform(df)
    df_scaled = pd.DataFrame(df_scaled, columns = df.columns)
    return df_scaled
```

```python
from sklearn.model_selection import train_test_split

df_scaled = feature_processing(df)  # (165034, 12)

X_train, X_test, y_train, y_test = train_test_split(df_scaled, train_target, test_size=0.3, random_state=42)
```

```python
df_scaled.head()
```

```python
from sklearn.linear_model import LinearRegression 

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_squared_error

mean_squared_error(y_test, model.predict(X_test), squared=True)  # 0.3580
```

```python
# model detail
print(model.coef_) # Partial Regression Coefficients
print(model.intercept_) # Intercept of the regression line
print(model.get_params()) # Get parameters
print(model.predict(X_test)) # Predicted values
print(model.score(X_test, y_test)) # coefficient of determination
```

```python
test_ids = data_test['id']

data_test_scaled = feature_processing(data_test)  # (110023, 12)

data_test_scaled.head()
```

```python
sub_target = model.predict(data_test_scaled)
print(sub_target[0])
```
